```python
import numpy as np
import matplotlib.pyplot as plt

#importanto o a camadad de Input do Keras
from keras.layers import Input, Dense, BatchNormalization

#Importando o modelo sequencial do Keras
from keras.models import Sequential
```

```python
totalFeatures = 6
windowSize = 20
num_hidden = 1500
```

```python
model = Sequential()

# Define a porta de entrada (o tamanho do input)
model.add(Input(shape=(totalFeatures*windowSize,)))

# 1. O Pulo do Gato: Batch Normalization logo de cara (igual ao nn.BatchNorm1d dela)
model.add(BatchNormalization())

# 2. As 6 camadas ocultas (Hidden Layers)
model.add(Dense(num_hidden, activation='relu'))
model.add(Dense(num_hidden, activation='relu'))
model.add(Dense(num_hidden, activation='relu'))
model.add(Dense(num_hidden, activation='relu'))
model.add(Dense(num_hidden, activation='relu'))
model.add(Dense(num_hidden, activation='relu'))

# 3. A camada de saída: 1 único neurônio, sem ReLU (ativação linear padrão)
# Isso equivale ao último nn.Linear(num_hidden, 1) do código dela
model.add(Dense(1))

model.summary()
```
