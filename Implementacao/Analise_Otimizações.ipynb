# ANÁLISE DAS OTIMIZAÇÕES

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## ETAPA 01

```python
Opt_exp_01 = pd.read_csv('Opt_exp_01.csv').sort_values(by='r2', ascending=False)
Opt_exp_01.head(10)
```

```python
# Configuração global de estilo para o Seaborn (deixa com cara de artigo acadêmico)
sns.set_theme(style="whitegrid", context="paper", font_scale=1.6)

# ==========================================
# GRÁFICO 1: Scatter Plot (Learning Rate vs R²)
# ==========================================
plt.figure(figsize=(8, 6))
sns.scatterplot(data=Opt_exp_01, x='learning_rate', y='r2', hue='function', palette='bright', s=100, alpha=0.8)
plt.xscale('log') 
plt.xlabel('Taxa de Aprendizado', fontweight='bold')
plt.ylabel('R²', fontweight='bold')
plt.legend(title='Função')
plt.tight_layout()
plt.show()


# ==========================================
# 1. HEATMAP: Camadas Ocultas vs Neurônios
# ==========================================
plt.figure(figsize=(8, 6))

heatmap_data = Opt_exp_01.groupby(['hiddenLayers', 'neurons'])['r2'].max().reset_index()
heatmap_pivot = heatmap_data.pivot(index='hiddenLayers', columns='neurons', values='r2')

sns.heatmap(heatmap_pivot, annot=True, fmt=".3f", linewidths=.5, cmap='Reds')
plt.xlabel('Número de Neurônios por Camada' , fontweight='bold')
plt.ylabel('Número de Camadas Ocultas' , fontweight='bold')
plt.gca().invert_yaxis() 
plt.tight_layout()
plt.show()

# ==========================================
# 2. BOXPLOTS: winSize e batchSize
# ==========================================
# Criando uma figura com dois gráficos lado a lado
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico da Esquerda: winSize
sns.boxplot(data=Opt_exp_01, x='winSize', y='r2', ax=axes[0], palette='pastel', width=0.5)
axes[0].set_xlabel('Tamanho da Janela ', fontweight='bold')
axes[0].set_ylabel('R²', fontweight='bold')

# Gráfico da Direita: batchSize
sns.boxplot(data=Opt_exp_01, x='batchSize', y='r2', ax=axes[1], palette='pastel', width=0.5)
axes[1].set_xlabel('Tamanho do Lote', fontweight='bold')
axes[1].set_ylabel('R²', fontweight='bold')

plt.tight_layout()
plt.show()
```
